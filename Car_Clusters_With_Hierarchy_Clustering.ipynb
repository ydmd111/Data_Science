# Importing the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the csv file:

```python
df=pd.read_csv('cluster_mpg.csv')
```

```python
df.head()
```

# To get the description:

```python
df.describe()
```

# To get the df info:

```python
df.info()
```

# To get each unique value counts in (origin) column :

```python
df['origin'].value_counts()
```

# Now we have to make dummies to the string columns with dropping the name column , because it's unique , so it's not useful in unsupervised learning :

```python
df_w_dummies=pd.get_dummies(df.drop('name',axis=1))
```

```python
df_w_dummies
```

# Then we call the MinMaxScaler library for scaling the data:

```python
from sklearn.preprocessing import MinMaxScaler
```

# Creating the object:

```python
scaler=MinMaxScaler()
```

# Then fit and transform the df_w_dummies :

```python
scaled_data=scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

# To create a data frame from the scaled_data:

```python
scaled_df=pd.DataFrame(scaled_data,columns=df_w_dummies.columns)
```

```python
scaled_df
```

# To draw the heatmap contains the scaled df:

```python
plt.figure(figsize=(15,8),dpi=200)
sns.heatmap(scaled_df,cmap='viridis')
```

# To use the heatmap with the default colors:

```python
plt.figure(figsize=(15,8),dpi=200)
sns.heatmap(scaled_df)
```

# Now let's use the clustermap :

```python
plt.figure(figsize=(15,8),dpi=200)
sns.clustermap(scaled_df,cmap='viridis')
```

# To draw the heatmap contains the correlation of scaled df:

```python
plt.figure(figsize=(15,8),dpi=200)
sns.heatmap(scaled_df.corr())
```

# Now let's import AgglomerativeClustering library for clustering:

```python
from sklearn.cluster import AgglomerativeClustering
```

# Then creat an object with 4 clusters :

```python
model=AgglomerativeClustering(n_clusters=4)
```

# Then fit and predict at the same step :

```python
cluster_labels=model.fit_predict(scaled_df)
```

```python
cluster_labels
```

# To plot the clusters :

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='mpg',y='weight',hue=cluster_labels,palette='viridis')
```

# Now let's import the hierarchy and dendrogram libraries :

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

# Creating a hierarchy object :

```python
linkage_matrix=hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

# Then draw the dendrogram with top 10 lines :

```python
plt.figure(figsize=(20,10))
dendro=dendrogram(linkage_matrix,truncate_mode='lastp',p=10)
```

# Now let's check the top 3 levels :

```python
plt.figure(figsize=(20,10))
dendro=dendrogram(linkage_matrix,truncate_mode='level',p=3)
```
