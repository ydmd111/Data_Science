# Now let's import the support vector machine library:

```python
from sklearn.svm import SVC
```

# Creating X & y values:

```python
X=df.drop('Virus Present',axis=1)
y=df['Virus Present']
```

# Creating the object and fit it:

```python
model=SVC(kernel='linear',C=1000)
```

```python
model.fit(X,y)
```

# Now let's import the plot_svm_boundary library:

```python
from svm_margin_plot import plot_svm_boundary
```

# To plot X & y with the use of model object:

```python
plot_svm_boundary(model,X,y)
```

# Now let's minimise the C value to 0.05 and fit it and plot it as previously:

```python
model=SVC(kernel='linear',C=0.05)
```

```python
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

# To get the best readings , let's import the GridSearchCV:

```python
from sklearn.model_selection import GridSearchCV
```

# And create an object for SVC:

```python
svm=SVC()
```

# Now let's put values for C and kernel in a dectionary:

```python
param_grid={'C':[0.01,0.1,1],'kernel':['linear','rbf']}
```

# Creating an object of GridSearch with svm object and param_grid:

```python
grid=GridSearchCV(svm,param_grid)
```

# And fit it:

```python
grid.fit(X,y)
```

# To get the best reading's faction in param_grid:

```python
grid.best_params_
```
