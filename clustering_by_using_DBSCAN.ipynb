# Importing the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the first csv file:

```python
moons=pd.read_csv('cluster_moons.csv')
moons.head()
```

# Reading the second csv file:

```python
circles=pd.read_csv('cluster_circles.csv')
circles.head()
```

# Then plot the moons infos :

```python
sns.scatterplot(data=moons,x='X1',y='X2')
```

# And plot the circles infos :

```python
sns.scatterplot(data=circles,x='X1',y='X2')
```

# Reading the third csv file:

```python
blobs=pd.read_csv('cluster_blobs.csv')
blobs.head()
```

# For ploting the blobs infos :

```python
sns.scatterplot(data=blobs,x='X1',y='X2')
```

# Then create a function to plot infos according to the chosen object :

```python
def display_categories(model,data):
    labels=model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

# Importing the DBSCAN clustering library :

```python
from sklearn.cluster import DBSCAN
```

# Creating an object :

```python
model=DBSCAN()
```

# Now let's apply this object on blobs :

```python
display_categories(model,blobs)
```

# Also let's apply this object on moons :

```python
display_categories(model,moons)
```

# Then to the circles :

```python
display_categories(model,circles)
```

# it's noticed that all infos in moons and circles are collected at the same cluster , to solve that we should make a new object which the epsolon less than default:

```python
model=DBSCAN(eps=0.15)
```

# Then apply the new object to moons :

```python
display_categories(model,moons)
```

# Also apply this new object on circles :

```python
display_categories(model,circles)
```
